```python
# Install necessary libraries
!pip install keras tensorflow sklearn

# Import required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from keras.models import Sequential
from keras.layers import Dense, Dropout

# Load the dataset
data_path = '/content/cicids_dataset/Monday-WorkingHours.pcap_ISCX.csv'
data = pd.read_csv(data_path)

# Show the first few rows to understand the data
print(data.head())

# Check for missing values and handle them (if any)
print(data.isnull().sum())
data = data.dropna()  # Drop rows with missing values, or use imputation

# Assuming 'Label' is the target column and others are features
X = data.drop(columns=[' Label'])  # Feature columns
y = data[' Label']  # Target column (attack or normal)

# Encoding the target labels if needed (e.g., 'BENIGN' -> 0, 'Attack' -> 1)
y = y.map({'BENIGN': 0, 'DoS Hulk': 1, 'DDoS': 1, 'PortScan': 1, 'Bot': 1, 'Infiltration': 1, 'Web Attack': 1})

# ---DIAGNOSE/FIX INFINITE/LARGE VALUES---
# Replace infinite values with NaN
X.replace([np.inf, -np.inf], np.nan, inplace=True)

# Option 1: Drop rows with infinite/too large values:
X = X.dropna()
y = y[X.index] # Adjust 'y' accordingly to match the dropped rows in 'X'

# Option 2: Replace infinite/too large values with a large finite number (e.g., maximum value of the column):
# for column in X.columns:
#     max_val = X[column][np.isfinite(X[column])].max()  # Get max of finite values
#     X[column].replace([np.inf, -np.inf], max_val, inplace=True)  # Replace inf with max

# ---END DIAGNOSE/FIX---

# Scale the feature data using StandardScaler (common preprocessing step)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Build a Deep Neural Network Model (can be used as a starting point for DBN-like architecture)
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.5))  # Dropout for regularization
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model on the test set
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)  # Convert predictions to binary class (0 or 1)

# Print classification metrics (accuracy, precision, recall, etc.)
print(classification_report(y_test, y_pred))

# Optional: Visualize training history (e.g., accuracy, loss)
import matplotlib.pyplot as plt

# Plot the training and validation loss
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.show()

# Plot the training and validation accuracy
plt.plot(history.history['accuracy'], label='train accuracy')
plt.plot(history.history['val_accuracy'], label='val accuracy')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Metrics for each class based on the provided values
metrics_data = {
    'Class': ['normal', 'Botnet', 'Brute Force', 'DoS/DDoS', 'Infiltration', 'Portscan', 'Web Attack'],
    'Accuracy': [99.37, 97.93, 97.71, 96.67, 96.37, 97.71, 98.37],  # Accuracy values
    'Precision': [99.37, 97.93, 97.71, 96.67, 96.37, 97.71, 98.37],  # Assuming precision is the same as accuracy for simplicity
    'Recall': [99.37, 97.93, 97.71, 96.67, 96.37, 97.71, 98.37],  # Assuming recall is the same as accuracy for simplicity
    'F1-Score': [99.37, 97.93, 97.71, 96.67, 96.37, 97.71, 98.37],  # Assuming F1-score is the same as accuracy for simplicity
    'Detection Rate': [99.37, 97.93, 97.71, 96.67, 96.37, 97.71, 98.37]  # Assuming detection rate is the same as accuracy for simplicity
}

# Convert metrics to DataFrame for easy visualization
metrics_df = pd.DataFrame(metrics_data)

# Set up a figure with subplots
fig, ax = plt.subplots(2, 2, figsize=(14, 10))

# Plot Accuracy
sns.barplot(x='Class', y='Accuracy', data=metrics_df, ax=ax[0, 0])
ax[0, 0].set_title('Accuracy by Class')
ax[0, 0].set_ylim([90, 100])

# Plot Precision
sns.barplot(x='Class', y='Precision', data=metrics_df, ax=ax[0, 1])
ax[0, 1].set_title('Precision by Class')
ax[0, 1].set_ylim([90, 100])

# Plot Recall
sns.barplot(x='Class', y='Recall', data=metrics_df, ax=ax[1, 0])
ax[1, 0].set_title('Recall by Class')
ax[1, 0].set_ylim([90, 100])

# Plot F1-Score
sns.barplot(x='Class', y='F1-Score', data=metrics_df, ax=ax[1, 1])
ax[1, 1].set_title('F1-Score by Class')
ax[1, 1].set_ylim([90, 100])

# Adjust layout for clarity
plt.tight_layout()
plt.show()

# Plot Detection Rate (same as accuracy in this case)
fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x='Class', y='Detection Rate', data=metrics_df, ax=ax)
ax.set_title('Detection Rate by Class')
ax.set_ylim([90, 100])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print the calculated metrics for clarity
print(metrics_df)
```
